## What is the average total queue time for water?

```python
%%sql
/* Average total queue time for water */
SELECT
    AVG(NULLIF(time_in_queue, 0)) AS total_avg_queue_time
FROM 
    visits;
```

## What is the average queue time on different days?

```python
%%sql
SELECT
    DAYNAME(time_of_record) AS day_of_week,
    ROUND(AVG(NULLIF(time_in_queue, 0))) AS avg_time
FROM 
    visits
GROUP BY 
    day_of_week;
```

## How can we communicate this information efficiently?
We can look at what time during the day people collect water. Try to order the results in a meaningful way.

```python
%%sql
/* Average queue time for different hours */
SELECT
    TIME_FORMAT(time_of_record, '%H:00') AS hour_of_day,
    ROUND(AVG(NULLIF(time_in_queue, 0))) AS avg_time
FROM 
    visits
GROUP BY 
    hour_of_day
ORDER BY 
    hour_of_day ASC;
```

Mornings and evenings are the busiest.It looks like people collect water before and after work. 

## Relaying this information according to hour of day for each day of the week.

```python
%%sql
SELECT
    TIME_FORMAT(TIME(time_of_record), '%H:00') AS hour_of_day,

    -- Sunday
    ROUND(AVG(
        CASE WHEN DAYNAME(time_of_record) = 'Sunday' THEN time_in_queue END
    ), 0) AS Sunday,

    -- Monday
    ROUND(AVG(
        CASE WHEN DAYNAME(time_of_record) = 'Monday' THEN time_in_queue END
    ), 0) AS Monday,

    -- Tuesday
    ROUND(AVG(
        CASE WHEN DAYNAME(time_of_record) = 'Tuesday' THEN time_in_queue END
    ), 0) AS Tuesday,

    -- Wednesday
    ROUND(AVG(
        CASE WHEN DAYNAME(time_of_record) = 'Wednesday' THEN time_in_queue END
    ), 0) AS Wednesday,

    -- Thursday
    ROUND(AVG(
        CASE WHEN DAYNAME(time_of_record) = 'Thursday' THEN time_in_queue END
    ), 0) AS Thursday,

    -- Friday
    ROUND(AVG(
        CASE WHEN DAYNAME(time_of_record) = 'Friday' THEN time_in_queue END
    ), 0) AS Friday,

    -- Saturday
    ROUND(AVG(
        CASE WHEN DAYNAME(time_of_record) = 'Saturday' THEN time_in_queue END
    ), 0) AS Saturday

FROM visits
WHERE time_in_queue != 0
GROUP BY hour_of_day
ORDER BY hour_of_day;
```

crews to drill for wells, providing a more permanent solution.
2. If communities are using wells, we can install filters to purify the water. For wells with biological contamination, we can install UV filters that

3. For shared taps, in the short term, we can send additional water tankers to the busiest taps, on the busiest days. We can use the queue time

queue times below 30 min.
4. Shared taps with short queue times (< 30 min) represent a logistical challenge to further reduce waiting times. The most effective solution,
installing taps in homes, is resource-intensive and better suited as a long-term goal.
5. Addressing broken infrastructure offers a significant impact even with just a single intervention. It is expensive to fix, but so many people
can benefit from repairing one facility. For example, fixing a reservoir or pipe that multiple taps are connected to. We will have to find the

## Is there a difference in the scores?
Comparing the scores in the auditor report and water quality tables

```python
%%sql
/* 
Investigating audit consistency by comparing:
- Auditor scores (true_water_source_score)
- Surveyor scores (subjective_quality_score)

We filter to visit_count = 1 to avoid duplicate entries for locations 
that were visited multiple times.
*/

SELECT
    auditor_report.location_id,
    auditor_report.true_water_source_score AS auditor_score,
    visits.record_id,
    water_quality.subjective_quality_score AS surveyor_score
FROM visits
INNER JOIN auditor_report 
    ON auditor_report.location_id = visits.location_id
INNER JOIN water_quality 
    ON visits.record_id = water_quality.record_id
WHERE auditor_report.true_water_source_score = water_quality.subjective_quality_score
  AND water_quality.visit_count = 1;

/* Use the SELECT statement below to return a simple confirmation message and avoid printing large result sets */
SELECT 'Updates complete' AS status;
```

## If so are there patterns?
After finding a difference in the scores from our first question we investigate the incorrect records with the above question

```python
%%sql
/* 
Checking auditor records that appear incorrect by identifying cases where:
- Auditor score ≠ Surveyor (subjective) score
We use visit_count = 1 to avoid duplicate records from multiple visits.
*/

SELECT
    auditor_report.location_id,
    auditor_report.true_water_source_score AS auditor_score,
    visits.record_id,
    water_quality.subjective_quality_score AS surveyor_score
FROM visits
INNER JOIN auditor_report 
    ON auditor_report.location_id = visits.location_id
INNER JOIN water_quality 
    ON visits.record_id = water_quality.record_id
WHERE auditor_report.true_water_source_score <> water_quality.subjective_quality_score
  AND water_quality.visit_count = 1;
```

```python
%%sql
/* 
Checking whether data entry errors originate from mismatched source types:
- Comparing auditor source type vs. surveyor source type
- Filtering only records where auditor vs. surveyor scores disagree
- visit_count = 1 ensures no duplicate visit entries
*/

SELECT
    auditor_report.location_id,
    auditor_report.type_of_water_source AS auditor_source,
    water_source.type_of_water_source AS surveyor_source,
    visits.record_id,
    auditor_report.true_water_source_score AS auditor_score,
    water_quality.subjective_quality_score AS surveyor_score
FROM visits
INNER JOIN auditor_report
    ON auditor_report.location_id = visits.location_id
INNER JOIN water_quality
    ON visits.record_id = water_quality.record_id
INNER JOIN water_source
    ON visits.source_id = water_source.source_id
WHERE auditor_report.true_water_source_score <> water_quality.subjective_quality_score
  AND water_quality.visit_count = 1;
```

**Types of sources look the same! So even though the scores are wrong, the integrity of the type_of_water_source

## Linking records to employees
Looking at where these errors may have come from. At some of the locations, employees assigned scores incorrectly, and those records
ended up in this results set.

Reasons this can happen:
1. These workers are all humans and make mistakes so this is expected.
2. Unfortunately, the alternative is that someone assigned scores incorrectly on purpose


```python
%%sql
/* 
Investigating whether audit errors might be linked to specific employees.
We compare auditor vs. surveyor scores and capture:
- Employee responsible
- Mismatched scores
- Auditor notes
Only visit_count = 1 is used to avoid duplicate records.
*/
CREATE VIEW Incorrect_records AS
SELECT
    auditor_report.location_id,
    visits.record_id,
    employee.employee_name,
    auditor_report.true_water_source_score AS auditor_score,
    water_quality.subjective_quality_score AS surveyor_score,
    auditor_report.statements AS statements
FROM visits
INNER JOIN auditor_report
    ON auditor_report.location_id = visits.location_id
INNER JOIN water_quality
    ON visits.record_id = water_quality.record_id
INNER JOIN employee
    ON visits.assigned_employee_id = employee.assigned_employee_id
WHERE auditor_report.true_water_source_score <> water_quality.subjective_quality_score
  AND water_quality.visit_count = 1;

/* SELECT line added at the end to avoid the resourceclosed error*/
SELECT 'Updates complete' AS status;
```

```python
%%sql
/* Identifying employees with above-average error counts */

WITH 
    error_count AS (
        SELECT 
            employee_name,
            COUNT(*) AS number_of_mistakes
        FROM Incorrect_records
        GROUP BY employee_name
    ),
    suspects_list AS (
        SELECT
            AVG(number_of_mistakes) AS avg_error_count_per_empl
        FROM error_count
    )
SELECT
    e.employee_name,
    e.number_of_mistakes
FROM error_count e
CROSS JOIN suspects_list s
WHERE e.number_of_mistakes > s.avg_error_count_per_empl;
```

- These are the employees who made more mistakes, on average, than their peers

- We should also look at the statements for these records to look for patterns

```python
%%sql
/* 
Investigating whether audit errors are linked to specific employees:
- Identify employees with above-average number of mistakes
- Retrieve all mismatched records for these employees
- visit_count = 1 ensures no duplicate visit records
- adding statements to investigate corruption patterns
*/

WITH Incorrect_records AS (
    SELECT
        auditor_report.location_id,
        visits.record_id,
        employee.employee_name,
        auditor_report.true_water_source_score AS auditor_score,
        water_quality.subjective_quality_score AS surveyor_score,
        auditor_report.statements AS statements
    FROM visits
    INNER JOIN auditor_report
        ON auditor_report.location_id = visits.location_id
    INNER JOIN water_quality
        ON visits.record_id = water_quality.record_id
    INNER JOIN employee
        ON visits.assigned_employee_id = employee.assigned_employee_id
    WHERE auditor_report.true_water_source_score <> water_quality.subjective_quality_score
      AND water_quality.visit_count = 1
),
error_count AS (
    SELECT 
        employee_name,
        COUNT(employee_name) AS number_of_mistakes
    FROM Incorrect_records
    GROUP BY employee_name
),
suspects_list AS (
    SELECT
        AVG(number_of_mistakes) AS avg_error_count_per_empl
    FROM error_count
),
above_avg_mistakes AS (
    SELECT
        e.employee_name,
        e.number_of_mistakes
    FROM error_count e
    CROSS JOIN suspects_list s
    WHERE e.number_of_mistakes > s.avg_error_count_per_empl
)
SELECT 
    i.employee_name,
    i.location_id,
    i.statements
FROM Incorrect_records i
JOIN above_avg_mistakes a
    ON i.employee_name = a.employee_name;
```

```python
%%sql
/* 
Creating a view `combined_analysis_table` that consolidates key data from multiple tables:
- water_source: type of source, number of people served
- location: town, province, and location type
- visits: time spent in queue
- well_pollution: contamination results
This simplifies downstream analysis by combining relevant information in a single view.
Only the first visit (visit_count = 1) is included to avoid duplicates.
*/

CREATE VIEW combined_analysis_table AS
SELECT
    water_source.type_of_water_source,
    location.town_name,
    location.province_name,
    location.location_type,
    water_source.number_of_people_served,
    visits.time_in_queue,
    well_pollution.results
FROM visits
LEFT JOIN well_pollution
    ON well_pollution.source_id = visits.source_id
INNER JOIN location
    ON location.location_id = visits.location_id
INNER JOIN water_source
    ON water_source.source_id = visits.source_id
WHERE visits.visit_count = 1;

SELECT 'Updates complete' AS status;
```

```python
%%sql

/* 
This query calculates the percentage of people served by each type of water source 
within each province. 

Steps:
1. `province_totals` CTE computes the total number of people served per province.
2. Main query calculates, for each province, the percentage of people served by:
   - river
   - shared_tap
   - tap_in_home
   - tap_in_home_broken
   - well
3. Percentages are rounded to whole numbers for easier interpretation.
*/

WITH province_totals AS (
    -- Calculate total population served per province
    SELECT
        province_name,
        SUM(number_of_people_served) AS total_ppl_serv
    FROM combined_analysis_table
    GROUP BY province_name
)

SELECT
    ct.province_name,
    -- Calculate percentage of population served by each water source type
    ROUND(
        (SUM(CASE WHEN type_of_water_source = 'river' THEN number_of_people_served ELSE 0 END) 
        * 100.0 / pt.total_ppl_serv), 0
    ) AS river,
    ROUND(
        (SUM(CASE WHEN type_of_water_source = 'shared_tap' THEN number_of_people_served ELSE 0 END) 
        * 100.0 / pt.total_ppl_serv), 0
    ) AS shared_tap,
    ROUND(
        (SUM(CASE WHEN type_of_water_source = 'tap_in_home' THEN number_of_people_served ELSE 0 END) 
        * 100.0 / pt.total_ppl_serv), 0
    ) AS tap_in_home,
    ROUND(
        (SUM(CASE WHEN type_of_water_source = 'tap_in_home_broken' THEN number_of_people_served ELSE 0 END) 
        * 100.0 / pt.total_ppl_serv), 0
    ) AS tap_in_home_broken,
    ROUND(
        (SUM(CASE WHEN type_of_water_source = 'well' THEN number_of_people_served ELSE 0 END) 
        * 100.0 / pt.total_ppl_serv), 0
    ) AS well
FROM combined_analysis_table ct
JOIN province_totals pt
    ON ct.province_name = pt.province_name
GROUP BY ct.province_name
ORDER BY ct.province_name;
```

```python
%%sql
/* 
Creating a temporary table `town_aggregated_water_access` to analyze water access by town:
1. `town_totals` CTE calculates the total number of people served per town, 
   using both province_name and town_name since town names may not be unique.
2. The main query calculates the percentage of people served by each type of water source per town:
   - river
   - shared_tap
   - tap_in_home
   - tap_in_home_broken
   - well
3. Percentages are rounded to whole numbers.
4. Join is performed on a composite key (province_name + town_name) to ensure correct matching.
*/

CREATE TEMPORARY TABLE town_aggregated_water_access
WITH town_totals AS (
    -- Calculate total population served per town
    SELECT 
        province_name, 
        town_name, 
        SUM(number_of_people_served) AS total_ppl_serv
    FROM combined_analysis_table
    GROUP BY province_name, town_name
)

SELECT
    ct.province_name,
    ct.town_name,
    -- Calculate percentages of population served by each water source type
    ROUND(
        (SUM(CASE WHEN type_of_water_source = 'river' 
            THEN number_of_people_served ELSE 0 END) 
        * 100.0 / tt.total_ppl_serv), 0
    ) AS river,
    ROUND(
        (SUM(CASE WHEN type_of_water_source = 'shared_tap' 
            THEN number_of_people_served ELSE 0 END) 
        * 100.0 / tt.total_ppl_serv), 0
    ) AS shared_tap,
    ROUND(
        (SUM(CASE WHEN type_of_water_source = 'tap_in_home' 
            THEN number_of_people_served ELSE 0 END) 
        * 100.0 / tt.total_ppl_serv), 0
    ) AS tap_in_home,
    ROUND(
        (SUM(CASE WHEN type_of_water_source = 'tap_in_home_broken' 
            THEN number_of_people_served ELSE 0 END) 
        * 100.0 / tt.total_ppl_serv), 0
    ) AS tap_in_home_broken,
    ROUND(
        (SUM(CASE WHEN type_of_water_source = 'well' 
            THEN number_of_people_served ELSE 0 END) 
        * 100.0 / tt.total_ppl_serv), 0
    ) AS well
FROM combined_analysis_table ct
-- Join on province + town composite key
JOIN town_totals tt 
    ON ct.province_name = tt.province_name
    AND ct.town_name = tt.town_name
-- Group by province and town to aggregate correctly
GROUP BY
    ct.province_name,
    ct.town_name
ORDER BY
    ct.town_name;

/* SELECT line added at the end to avoid the resourceclosed error*/
SELECT 'Updates complete' AS status;
```

## PROJECT PROGRESS
This is in allignment with the project plan.We create a project progress table and add relevant information to it.

```python
%%sql
/* 
Creating the `Project_progress` table to track improvements on water sources:
- `Project_id`: unique identifier for each project (auto-incremented)
- `source_id`: links to `water_source` table, with cascading updates/deletes
- `Address`, `Town`, `Province`: location details
- `Source_type`: type of water source
- `Improvement`: planned improvement/action
- `Source_status`: current status of the project, constrained to Backlog, In progress, or Complete
- `Date_of_completion`: expected or actual completion date
- `Comments`: additional notes
*/

CREATE TABLE Project_progress (
    Project_id SERIAL PRIMARY KEY,
    source_id VARCHAR(20) NOT NULL 
        REFERENCES water_source(source_id) 
        ON DELETE CASCADE 
        ON UPDATE CASCADE,
    Address VARCHAR(50),
    Town VARCHAR(30),
    Province VARCHAR(30),
    Source_type VARCHAR(50),
    Improvement VARCHAR(50),
    Source_status VARCHAR(50) DEFAULT 'Backlog' 
        CHECK (Source_status IN ('Backlog', 'In progress', 'Complete')),
    Date_of_completion DATE,
    Comments TEXT
);

SELECT 'Updates complete' AS status;
```

```python
%%sql
/*
Selecting water sources that may require attention or improvements. 

Criteria for inclusion:
1. Only the first visit per location is considered (visit_count = 1) to avoid duplicates.
2. Any of the following conditions must be true:
   - The water source is marked as polluted (well_pollution.description != 'Clean')
   - The water source is a broken tap or a river (potentially unreliable sources)
   - The water source is a shared tap with a long queue time (>= 30 minutes)

Columns returned:
- Address, Town, Province: location details
- source_id, type_of_water_source: water source details
- results: pollution/contamination results from well_pollution table
*/

SELECT
    location.address,
    location.town_name,
    location.province_name,
    water_source.source_id,
    water_source.type_of_water_source,
    well_pollution.results
FROM water_source
LEFT JOIN well_pollution 
    ON water_source.source_id = well_pollution.source_id
INNER JOIN visits 
    ON water_source.source_id = visits.source_id
INNER JOIN location 
    ON location.location_id = visits.location_id
WHERE 
    visits.visit_count = 1
    AND (
        well_pollution.description != 'Clean'
        OR water_source.type_of_water_source IN ('tap_in_home_broken', 'river')
        OR (water_source.type_of_water_source = 'shared_tap' AND visits.time_in_queue >= 30)
    );
```

```python
%%sql
/*
This query generates a recommended improvement plan for water sources based on their type, 
pollution status, and service conditions.

Columns:
- Address, Town, Province: location details
- source_id, Source_type: water source details
- Source_status: contamination or pollution results
- Improvement: recommended action based on conditions

Logic for Improvement recommendations:
1. If pollution results indicate chemical contamination → "Install RO filter"
2. If pollution results indicate biological contamination → "Install UV and RO filter"
3. If source is a river → "Drill Well"
4. If source is a shared tap with long queue time (>=30 mins) → Install additional taps
5. If source is a broken tap in home → "Diagnose Local Infrastructure"
6. Else → "No Improvement Needed"

Only the first visit (visit_count = 1) is considered to avoid duplicates. 
Sources are included if any of the following is true:
- Water is not clean
- Source is a river or broken home tap
- Shared tap with long queue time
*/

SELECT
    location.address AS Address,
    location.town_name AS Town,
    location.province_name AS Province,
    water_source.source_id,
    water_source.type_of_water_source AS Source_type,
    well_pollution.results AS Source_status,

    CASE
        WHEN well_pollution.results LIKE "%Chemical%" 
            THEN "Install RO filter"

        WHEN well_pollution.results LIKE "%Biological%" 
            THEN "Install UV and RO filter"

        WHEN water_source.type_of_water_source = "river"
            THEN "Drill Well"

        WHEN water_source.type_of_water_source = "shared_tap"
             AND visits.time_in_queue >= 30
            THEN CONCAT("Install ", FLOOR(visits.time_in_queue/30), " tap(s) nearby")

        WHEN water_source.type_of_water_source = "tap_in_home_broken"
            THEN "Diagnose Local Infrastructure"

        ELSE "No Improvement Needed"
    END AS Improvement

FROM water_source
LEFT JOIN well_pollution 
    ON water_source.source_id = well_pollution.source_id
INNER JOIN visits 
    ON water_source.source_id = visits.source_id
INNER JOIN location 
    ON location.location_id = visits.location_id

WHERE 
    visits.visit_count = 1 
    AND (
            well_pollution.description != 'Clean'
         OR water_source.type_of_water_source IN ('tap_in_home_broken','river')
         OR (water_source.type_of_water_source = 'shared_tap' AND visits.time_in_queue >= 30)
        );
```

```python
%%sql
/*
Inserting records into `Project_progress` based on analysis of water sources. 

Columns populated:
- Address, Town, Province: location details
- source_id, Source_type: water source identifiers
- Improvement: recommended action based on water source type, pollution, and service conditions

Logic for Improvement:
1. Chemical contamination → "Install RO filter"
2. Biological contamination → "Install UV and RO filter"
3. River sources → "Drill Well"
4. Shared taps with long queue (>=30 mins) → Install additional taps
5. Broken home taps → "Diagnose Local Infrastructure"
6. Otherwise → "No Improvement Needed"

Filters:
- Only the first visit per location (visit_count = 1) is considered
- Only sources meeting one of the conditions below are included:
   a) Pollution detected (description != 'Clean')
   b) Broken taps or rivers
   c) Shared taps with long queues
*/

INSERT INTO Project_progress (
    Address,
    Town,
    Province,
    source_id,
    Source_type,
    Improvement
)
SELECT
    location.address AS Address,
    location.town_name AS Town,
    location.province_name AS Province,
    water_source.source_id,
    water_source.type_of_water_source AS Source_type,
    CASE
        WHEN well_pollution.results LIKE "%Chemical%" 
            THEN "Install RO filter"
        WHEN well_pollution.results LIKE "%Biological%" 
            THEN "Install UV and RO filter"
        WHEN water_source.type_of_water_source = "river"
            THEN "Drill Well"
        WHEN water_source.type_of_water_source = "shared_tap"
             AND visits.time_in_queue >= 30
            THEN CONCAT("Install ", FLOOR(visits.time_in_queue/30), " tap(s) nearby")
        WHEN water_source.type_of_water_source = "tap_in_home_broken"
            THEN "Diagnose Local Infrastructure"
        ELSE "No Improvement Needed"
    END AS Improvement
FROM water_source
LEFT JOIN well_pollution 
    ON water_source.source_id = well_pollution.source_id
INNER JOIN visits 
    ON water_source.source_id = visits.source_id
INNER JOIN location 
    ON location.location_id = visits.location_id
WHERE 
    visits.visit_count = 1 
    AND (
            well_pollution.description != 'Clean'
         OR water_source.type_of_water_source IN ('tap_in_home_broken','river')
         OR (water_source.type_of_water_source = 'shared_tap' AND visits.time_in_queue >= 30)
        );

/* SELECT line added at the end to avoid the resourceclosed error*/
SELECT 'Updates complete' AS status;
```
